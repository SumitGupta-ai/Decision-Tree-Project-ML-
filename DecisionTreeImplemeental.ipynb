### Load Iris Dataset 
##### This is a famous machine learning dataset commonly used for classification problems.


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.target
```

```python
df = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)

# Target (label) column adding
df['target'] = iris.target
```

```python
df
```

```python
# Independent Feature and Dependent Feature
x = df.drop('target', axis=1)
y = df['target']
```

```python
x.head(3)
```

```python
y.head(3)
```

### Train Test Split

```python
# Train test split data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

### DecisionTree Algorithm (Apply)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(max_depth=2)
```

```python
dt_model.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(dt_model, filled=True)
plt.show()
```

### Prediction and Evalution Matrics

```python
# Prediction
y_pred = dt_model.predict(x_test)
```

```python
print('Prediction Test Data' ,y_pred)
```

```python
from sklearn.metrics import  accuracy_score, classification_report, confusion_matrix
```

```python
# Model Accuracy
score = accuracy_score(y_test, y_pred)
print(f'Accuracy score: {score:.3f}')
```

```python
print(classification_report(y_pred, y_test))
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print('Confusion Matix\n', cm)
```

## After Using Decision Tree Parameters:

```python
parameter = {
    "criterion" : ["gini", "entropy", "log_loss"],
    'splitter' :["best", "random"],
    'max_depth' : [1,2,3,4,5],
    'max_features' :['sqrt', 'log2', 'auto']
}
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
tree_model = DecisionTreeClassifier()
cv = GridSearchCV(tree_model, param_grid=parameter, scoring='accuracy', cv=5)
```

```python
cv.fit(x_train, y_train)
```

### Best Parameter in Dataset

```python
cv.best_params_
```

```python
y_pred = cv.predict(x_test)
```

```python
y_pred
```

```python
print('accuracy socre:', accuracy_score(y_test, y_pred))
```

```python

print('Clasification Report:', classification_report(y_test, y_pred))
```

```python
print(cv.best_score_)
```

```python
print(confusion_matrix(y_test, y_pred))
```
